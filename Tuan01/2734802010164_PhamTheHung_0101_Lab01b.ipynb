# 6. ghp nh ngang v dc 2 nh

```python
img_r1 = cv.resize(img, (300, 300))
img_r2 = cv.resize(img2, (300, 300))
```

ghp ngang

```python
img_h = cv.hconcat([img_r1, img_r2])

PhamTheHung.imshow(cv.cvtColor(img_h, cv.COLOR_BGR2RGB))
PhamTheHung.axis('off')
```

```python
img_v = cv.vconcat([img_r1, img_r2])

PhamTheHung.imshow(cv.cvtColor(img_v, cv.COLOR_BGR2RGB))
PhamTheHung.axis('off')
```
