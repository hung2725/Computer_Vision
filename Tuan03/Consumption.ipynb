# Introduction to the Uber Dataset

```python
import pandas as pd
#Read in the dataset of july uber pickups
uber_data = pd.read_csv("uber-raw-data-jul14.csv")
print(uber_data.tail(10))
uber_data.info()
```

```python
#Convert the type of the column to datetime
uber_data["Date/Time"] = pd.to_datetime(uber_data["Date/Time"])
```

```python
#Using a floor can round off the date-time into discrete increments
print(uber_data["Date/Time"].dt.floor('1h').tail(10))
```

```python
#Let's get value counts to see the number of trips at given times
print(uber_data["Date/Time"].dt.floor('1h').value_counts().head(10))
```

```python
#And sort the data so it is chronological
hourly_data = uber_data["Date/Time"].dt.floor('1h').value_counts()
hourly_data = hourly_data.sort_index()
print(hourly_data.head(10))
```

```python
import matplotlib.pyplot as plt
#Plotting the data shows some trend components
hourly_data.plot(kind="line")
plt.show()
```

```python
#We are going to see the average number of trips for each hour/week day combination
#First split the date into the week day, hour and the actual date
hours = uber_data["Date/Time"].dt.hour# trch xut gi
week_day = uber_data["Date/Time"].dt.weekday# trch xut ngy trong tun (0=Th Hai, 6=Ch Nht)
date = uber_data["Date/Time"].dt.date# trch xut ngy thng nm
weekly_data = pd.concat([week_day, hours, date], axis=1) # ghp cc Series thnh DataFrame
weekly_data.columns = ["Week Day", "Hour", "Date"] # t tn cho cc ct
print(week_day.iloc[66003])# in ra gi tr ngy trong tun ca dng th 66003
print(weekly_data.tail(10)) # in ra 10 dng cui ca DataFrame weekly_data
```

```python
import calendar
#The calendar library can map the integer versions of calendar weekdays to the actual name
#0 -> Monday, 1 -> Tuesday, etc.
print(calendar.day_name[6])# in ra tn ngy tng ng vi s 6
```

```python
#Map the name
weekly_data["Week Day"] = weekly_data["Week Day"].apply(lambda x: calendar.day_name[x]) # nh x s ngy trong tun thnh tn ngy
print(weekly_data["Week Day"].iloc[66003])
```

```python
#By grouping by the date, week day, and hour we can aggregate the size (# of entries) on each date
weekly_data = weekly_data.groupby(["Date","Week Day", "Hour"]).size() # nhm theo ngy, ngy trong tun v gi, sau  m s lng mc trong mi nhm
print(weekly_data.head(10))
```

```python
#Reset the index
weekly_data = weekly_data.reset_index()# t li ch mc
print(weekly_data.head(10))
```

```python
#Rename 0, the default column name to be size
weekly_data = weekly_data.rename(columns={0: "Size"})# i tn ct mc nh 0 thnh Size
print(weekly_data.head(10))
```

```python
#Now we can group by the week day and average to get the mean for each week day/hour
weekly_data = weekly_data.groupby(["Week Day", "Hour"])["Size"].mean() # nhm theo ngy trong tun v gi, sau  tnh trung bnh s lng mc trong mi nhm
print(weekly_data.head(10))
```

```python
#Unstack takes a level of the index and translates it to be a column
#We pick level=0 because we want the week day name to be the column
weekly_data = weekly_data.unstack(level=0) # chuyn i cc ngy trong tun thnh ct
print(weekly_data)
```

```python
#Reindex allows you to re-arrange the columns however you would like
# sp xp li cc ct theo th t mong mun
weekly_data = weekly_data.reindex(columns=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
print(weekly_data)
```

```python
import seaborn as sns
#Plot a heatmap of the data
sns.heatmap(weekly_data)
plt.show()
```

```python
#Change the color map to blue
sns.heatmap(weekly_data, cmap='Blues')
plt.show()
```

# Points of Interest

```python
#geopy is a library which finds distance between latitude and longitude! python3 -m 
%pip install geopy
import geopy.distance
```

```python
#Check to make sure latitude and longitude are in the right order
metro_art_coordinates = (40.7794, -73.9632)# ta  ca Vn Lang
empire_state_building_coordinates = (40.7484, -73.9857)
distance = geopy.distance.distance(metro_art_coordinates, empire_state_building_coordinates) # tnh khong cch gia hai a im
print(distance)
print(distance.mi)
```

```python
#Easy way to convert our latitude and longitude columns to tuples
print(uber_data[["Lat", "Lon"]].apply(lambda x: tuple(x),axis=1))# chuyn i cc ct Lat v Lon thnh cc tuple (Lat, Lon) cho mi hng
```

```python
#Using the geopy version may take too long, so we will use the haversine formula instead
from math import radians, cos, sin, asin, sqrt

def haversine(coordinates1, coordinates2):# hm tnh khong cch gia hai ta  a l s dng cng thc haversine

    lon1 = coordinates1[1]
    lat1 = coordinates1[0]
    lon2 = coordinates2[1]
    lat2 = coordinates2[0]
    #Change to radians
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])# chuyn i  sang radian
    
    
    # Apply the harversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2# tnh ton theo cng thc haversine
    c = 2 * asin(sqrt(a)) 
    r = 3956
    return c * r
print(haversine(metro_art_coordinates, empire_state_building_coordinates))# in ra khong cch gia hai a im s dng cng thc haversine
print(distance.mi)
```

```python
#Now, we can find the distances to both attractions
# tnh khong cch t v tr ca Uber n Vn Lang
uber_data["Distance MM"] = uber_data[["Lat", "Lon"]].apply(lambda x: haversine(metro_art_coordinates,tuple(x)),axis=1) 
# tnh khong cch t v tr ca Uber n Empire State Building
uber_data["Distance ESB"] = uber_data[["Lat", "Lon"]].apply(lambda x: haversine(empire_state_building_coordinates,tuple(x)),axis=1)
print(uber_data["Distance MM"].head(5))
print(uber_data["Distance ESB"].head(5))
```

```python
#Summarize the data
print(uber_data[["Distance MM", "Distance ESB"]].describe())# tm tt cc thng k m t ca hai ct khong cch
```

```python
#Using boolean indexing, we can sum to find the count within a specified range
print((uber_data[["Distance MM", "Distance ESB"]] < .25).sum()) # m s lng cc chuyn i c khong cch n Vn Lang v Empire State Building nh hn 0.25 dm
```

```python
import numpy as np
#Distance range takes a start, end (non-inclusive) and step amount
distance_range = np.arange(.1,5.1,.1) # to mng khong cch t 0.1 n 5.0 vi bc nhy 0.1
print(distance_range)
```

```python
#Run our analysis for each distance
# phn tch s lng chuyn i trong mi khong cch xc nh
distance_data = [(uber_data[["Distance MM", "Distance ESB"]] < dist).sum() for dist in distance_range]
print(distance_data)
```

```python
#Concat
distance_data = pd.concat(distance_data, axis=1) # ghp cc Series trong distance_data thnh mt DataFrame
print(distance_data)
```

```python
#Transpose and add in the index
distance_data = distance_data.transpose()# chuyn v DataFrame  cc khong cch tr thnh ch mc
distance_data.index = distance_range
print(distance_data)
```

```python
#And plot
distance_data.plot(kind="line")# v biu  ng th hin s lng chuyn i trong mi khong cch
plt.show()
```

```python
#Let's grab only the date and hour by replacing the other parts with 0
# to ct Date_Hour bng cch thay th pht, giy v micro giy trong ct Date/Time thnh 0
uber_data["Date_Hour"] = uber_data["Date/Time"].apply(lambda x: x.replace(microsecond=0,second=0,minute=0))
print(uber_data["Date_Hour"])
```

```python
from datetime import datetime
#Take only the first week of data
map_data = uber_data[uber_data["Date/Time"] < datetime(2014,7,8)].copy()# lc d liu Uber  ch gi li cc chuyn i trong tun u tin ca thng 7 nm 2014
map_data["Weight"] = .5 # thm ct Weight vi gi tr c nh 0.5 cho mi hng
#Randomly sample 1/3 the values in each group
map_data = map_data.groupby('Date_Hour').apply(lambda x: x[["Lat", "Lon", "Weight"]].sample(int(len(x)/3)).values.tolist())# nhm theo ct Date_Hour v ly mu ngu nhin 1/3 cc gi tr trong mi nhm
#Get the index
date_hour_index = [x.strftime("%m/%d/%Y, %H:%M:%S") for x in map_data.index] # nh dng li ch mc thnh chui ngy gi
#Get the data in list form (each element of this bigger list will be a list of lists with lat/lon/weight)
#Each element of the bigger list is a for a date/hour combo
date_hour_data = map_data.tolist()
```

```python
from folium.plugins import HeatMapWithTime
# Thay i t ta  c sang ta  i hc Vn Lang
uber_map = folium.Map(location=[10.827489, 106.700059], zoom_start=15)  # to bn  vi ta  trung tm v mc thu phng c th
#A heatmap with time can now be out together
hm = HeatMapWithTime(date_hour_data, index=date_hour_index)# to bn  nhit theo thi gian vi d liu v ch mc tng ng
hm.add_to(uber_map)
uber_map
```

# Testing Seasonality

```python
#Recall the seasonality we saw before
hourly_data.plot(kind='line')
plt.show()
```

```python
#What about the hourly trends?
h = hourly_data.groupby(hourly_data.index.hour).mean()
h.plot(kind="line")
plt.show()
```

```python
#Something else of interest is the difference in hourly trends for weekdays and weekends
#We will index with i for weekdays
i = hourly_data.index.weekday <= 4

h_week = hourly_data.loc[i].groupby(hourly_data.loc[i].index.hour).mean()
h_weekend = hourly_data.loc[~i].groupby(hourly_data.loc[~i].index.hour).mean()
h = pd.concat([h_week, h_weekend], axis=1)
h.columns = ["Weekday", "Weekend"]
print(h)
```

```python
#And plot to see the difference
h.plot(kind='line')
plt.show()
```

```python
#We can also divide by the total number of trips for each to normalize and have each be a percent of total trips in a day
(h / h.sum()).plot(kind='line')
plt.show()
```

```python
#We can finish our assessment of whether or not we see hourly effects by using a t-test to see if each hour
#has a statistically different proportion of rides for weekends vs. weekdays
from scipy.stats import ttest_ind

#The functions takes two samples and returns the t-stat and the p-value denoting the null hypothesis that they are the same
print(ttest_ind([100,105,110], [200,230,210]))
```

```python
#Now for each day, let's normalize by the total number of rides in the day
hourly_data_pct = hourly_data.groupby(hourly_data.index.date).apply(lambda x: x / x.sum())
print(hourly_data_pct)
```
